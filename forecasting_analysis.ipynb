# Financial Revenue Forecasting 

This notebook builds a simple forecasting model to predict future revenues for Microsoft, Tesla and Apple using historical Form 10-K data.

```python
import pandas as pd 
import matplotlib.pyplot as plt 

df = pd.read_excel('3 yr financials.xlsx')
df
```

We use linear regression to establish a baseline forecasting model.
While simple, it provides an interpretable benchmark for future improvements. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error, r2_score

data = {
    'year': [2022, 2023, 2024],
    'revenue': [1998270, 211915, 245122]
}
df = pd.DataFrame(data)
df = df.sort_values(by='year')
print("Sorted Data:")
print(df)
x = df['year'].values.reshape(-1, 1)
y = df['revenue'].values 
model = LinearRegression()
model.fit(x, y)
y_pred = model.predict(x)

plt.scatter(df['year'], df['revenue'], color='blue', label='Actual data')
plt.plot(df['year'], y_pred,color='green', label='Regression line')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.title('Linear Regression of Revenue vs Year')
plt.legend()
plt.show()
```

## Forecasting Assumption 

The forecasting model is based on is based on the following assumption:
-Revenue follows a linear trend overtime.
-No major structural changes occur in the company's market or operations.
-Historical revenue perfomance is indicative of short-term future trends. 

```python
import numpy as np
import matplotlib.pyplot as plt
future_years = np.array([[2025], [2026]])
future_predictions = model.predict(future_years)
plt.scatter(future_years,future_predictions, color='red', label='Predictions') 

plt.show()
```

## Model Limitation

Thid model is trained on a very small dataset ( three observations), which limits its predictive reliability. No train-test split or cross-validation was perfomed due to data constraints. The forecast should be interpreted as an illustrative baseline rather than a precise prediction.
